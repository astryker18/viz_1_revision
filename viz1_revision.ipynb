```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/Users/Alex/Google Drive/Current Classes/Advanced Composing of Digital Media/Final Project/calendarcopy.csv')
df1 = pd.DataFrame(data)

df2 = df1[:15]
df3 = df1[15:]
```

```python
fig, ax = plt.subplots(figsize = (30, 12))

pos = [y*50 for y in list(range(len(df2.index)))]

plt.bar(pos, df2["Alpha Phi Omega"], width=5, color='cyan', edgecolor='black')
plt.bar([i+5 for i in pos], df2["Classes"], width=5, color='gray', edgecolor='black')
plt.bar([i+10 for i in pos], df2["Work"], width=5, color='red', edgecolor='black')
plt.bar([i+15 for i in pos], df2["Homework"], width=5, color='purple', edgecolor='black')
plt.bar([i+20 for i in pos], df2["Meals"], width=5, color='orange', edgecolor='black')
plt.bar([i+25 for i in pos], df2["Morning Routine"], width=5, color='green', edgecolor='black')
plt.bar([i+30 for i in pos], df2["Sleep"], width=5, color='yellow', edgecolor='black')
plt.bar([i+35 for i in pos], df2["Miscellaneous"], width=5, color='blue', edgecolor='black')

plt.plot([], color='cyan', label='Alpha Phi Omega', linewidth=5)
plt.plot([], color='gray', label='Classes', linewidth=5)
plt.plot([], color='red', label='Work', linewidth=5)
plt.plot([], color='purple', label='Homework', linewidth=5)
plt.plot([], color='orange', label='Meals', linewidth=5)
plt.plot([], color='green', label='Morning Routine', linewidth=5)
plt.plot([], color='yellow', label='Sleep', linewidth=5)
plt.plot([], color='blue', label='Miscellaneous', linewidth=5)

plt.legend(loc=2, ncol=2, prop={'size': 13}, edgecolor='black', facecolor='white')

ax.set_xticklabels(df1["Day"])
ax.set_xticks([p + 2.8 * 8 for p in pos])

plt.xlabel('Day Number', fontsize=15)
plt.ylabel('Hours Per Day', fontsize=12)
plt.title('Time Spent on Activities Each Day, Days 1-15', fontsize=16, weight="bold")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("Revision1.png")
plt.show()
```
